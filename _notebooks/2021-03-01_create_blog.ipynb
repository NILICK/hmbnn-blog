# Building a blog with Fastpages
1. For create and check blog posts in *localhost*, it needs dicker-compose.
For install it in Ubuntu first in base environment (src=https://docs.docker.com/compose/install/):

```
sudo curl -L "https://github.com/docker/compose/releases/download/1.28.5/docker-compose-$(uname -s)-$(uname -m)" -o /usr/local/bin/docker-compose
```

2. Next for permissions:
```
sudo chmod +x /usr/local/bin/docker-compose
```
3. Now create a repository in *github* for build *fastpages* blog:
- https://github.com/fastai/fastpages/generate
4. rep name = hmblog
5. create repository from template
6. in new rep created after a few moments, click on *Pull requests* and then *initial Setup*
7. create an ssh key-pair from:https://8gwifi.org/sshfunctions.jsp
8. with *RSA* and *4096* options Generate-SSH-Keys
9. copy all text in *Private Key* and pate it in Action *Secret* of own rep with name of *SSH_DEPLOY_KEY*
10. copy all text in *Public Key* and paste it in *Deploy keys* and check *Allow write access* with name of *blog-key*.
11. go back to *Initial Setup* page and in pagedown click on the *Merge pull request* and then *Confirm request*.
12. after a few moments in *code* page click on url="https://xxx.github.io/hbnn/" and see the preliminary blog.
13. *_config.yml* is for change some metadata such as title, date, category,...
14. for edit this file open it and change desired texts then commit changes.

---------------------------------------------------------------------------------------------------------
BECAUSE THERE ARE SOME PROBLEMS IN WORK WITH DOCKER IN LOCALHOST IT IS BETTER DONE ANY CHANGES IN GITHUB WEBSITE!

15. because of fast blogging it is best choice to run in localhost and then upload completed blog posts.
16. for do it first install "docker engine" and "docker-compose"
17. for install docker:
```
		# DOCKER ENGINE
		sudo apt-get remove docker docker-engine docker.io containerd runc
		sudo apt-get update
		sudo apt-get install \
		    apt-transport-https \
			ca-certificates \
			curl \
			gnupg \
			lsb-release
			
		curl -fsSL https://download.docker.com/linux/ubuntu/gpg | sudo gpg --dearmor -o /usr/share/keyrings/docker-archive-keyring.gpg
		
		echo \
		"deb [arch=amd64 signed-by=/usr/share/keyrings/docker-archive-keyring.gpg] https://download.docker.com/linux/ubuntu \
		(lsb_release -cs) stable" | sudo tee /etc/apt/sources.list.d/docker.list > /dev/null
		
		# FOR VERIFY INSTALLED DOCKER ENGINE
		sudo docker run hello-world
		
		# DOCKER-COMPOSE
		sudo curl -L "https://github.com/docker/compose/releases/download/1.28.5/docker-compose-$(uname -s)-$(uname -m)" -o /usr/local/bin/docker-compose
		sudo chmod +x /usr/local/bin/docker-compose
		
```
18. for do it, we need docker and first created repository must be downloaded in local drive.
19. after download and unzip the repository, open terminal in root of local repository.
20. sudo make build # (this is for fist time only)
21. sudo make server
22. open blog created locally with rep name: http://0.0.0.0:4000/hbnn/
23. for stop server "ctrl + c"
24. EDIT "_config.yml" for show an image side of post title:

		show_image: true
		
25. Save "_config.yml" and 
	sudo make server

(for more reading: https://docs.docker.com/config/pruning/ & https://www.digitalocean.com/community/tutorials/how-to-remove-docker-images-containers-and-volumes)	
26. for cleanup all images, containers and volumes locally use:
	*list of all images*
	sudo docker images -a
	*list of container*
	sudo docker container ls
	*list of volumes*
	sudo docker volume ls
	
	*Delete anything*
	docker system prune
	docker system prune --volumes
	docker image prune -a

# Github Desktop Step by Step

1. Download last relased Github Desktop deb format from:
https://github.com/shiftkey/desktop/releases
2. Open terminal and run update commandas:
sudo apt update && sudo apt install -f
3. Install Github Desktop package:
sudo dpkg -i fileName.deb
4. After install open package in terminal with:
Github desktop
5. Getting started with Github Desktop:
Open GitHub Desktop. Click ‘Sign in to GitHub.com’ 
6. Sign in using your GitHub username & password
7. Update your user information if necessary
8. Click ‘Finish’
9. Create new repository OR Clone available repository in github.com to local github desktop directory.
10. With add any file(s) or edit any availe files in github repository, it will be show in github desktop in left sidebar changes tab.
11. To push the changes in github, you must first type a comment as description in *summary (required)* in left sidebar and then commit to master.
12. After commited changes, now it is time to *Push origin*.
13. Now you can check your github's repository and can see changes.
